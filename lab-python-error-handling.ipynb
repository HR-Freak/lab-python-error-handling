````python
products = ["t-shirt", "mug", "hat", "book", "keychain"]
````

````python
def initialize_inventory(products):
    inventory = {}
    for product in products:
        valid_quantity = False
        while not valid_quantity:
            try:
                quantity = int(input(f"Enter the quantity of {product}s available: "))
                if quantity < 0:
                    raise ValueError("Invalid quantity! Please enter a non-negative value.")
                valid_quantity = True
            except ValueError as error:
                print(f"Error: {error}")
        inventory[product] = quantity
    return inventory
````

````python
inventory = initialize_inventory(products)
print("Initial inventory:", inventory)
````

````python
def get_customer_orders(products, inventory):
    while True:
        try:
            num_orders = int(input("How many different products would you like to order? "))
            if num_orders <= 0:
                raise ValueError("Number of products must be positive.")
            break
        except ValueError as e:
            print(f"Invalid input: {e}. Please enter a positive integer.")

    customer_orders = set()
    for i in range(num_orders):
        while True:
            product_name = input(f" Enter the name of product {i + 1}: ")
            try:
                if product_name not in products:
                    raise ValueError("Product not found.")
                if inventory.get(product_name, 0) <= 0:
                    raise ValueError(f"Product out of stock.")
                customer_orders.add(product_name)
                break
            except ValueError as e:
                print(f"Invalid input: {e}.")
    return customer_orders
````

````python
orders = get_customer_orders(products, inventory)
print("Customer orders:", orders)
````

````python
def calculate_total_price(customer_orders):
    total_price = 0.0
    for p in customer_orders:
        while True:
            raw = input(f"Enter the price for {p}: ")
            try:
                price = float(raw)
                if price < 0:
                    raise ValueError("Price cannot be negative.")
                total_price += price
                break
            except ValueError as e:
                print(f"Invalid input: {e}. Please enter a valid price.")
    return total_price
````

````python
total_price = calculate_total_price(orders)
print(f"Total price of the order: €{total_price}")
````
